```python
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.common.keys import Keys
import json
urls=[
    'https://bachelors.nl/zoeken/?orgUnitId=&programForm=full-time&profile=&studyCluster=economics&programType=&searchWord=',
    'https://bachelors.nl/zoeken/?orgUnitId=&programForm=full-time&profile=&studyCluster=behaviour&programType=&searchWord=',
    'https://bachelors.nl/zoeken/?orgUnitId=&programForm=full-time&profile=&studyCluster=health&programType=&searchWord=',
    'https://bachelors.nl/zoeken/?orgUnitId=&programForm=full-time&profile=&studyCluster=agriculture&programType=&searchWord=',
    'https://bachelors.nl/zoeken/?orgUnitId=&programForm=full-time&profile=&studyCluster=nature&programType=&searchWord=',
    'https://bachelors.nl/zoeken/?orgUnitId=&programForm=full-time&profile=&studyCluster=law&programType=&searchWord=',
]
clusters=[
    'Economie',
    'Gedrag en maatschappij',
    'Gezondheidszorg',
    'Landbouw en natuurlijke omgeving',
    'Natuur',
    'Recht',
]  
dict_list=[]
# problem_numbers=[]
for i in range(6):
    # university=universities[i]
    # degree=degrees[i]

    cluster=clusters[i] 
    url=urls[i]
    

    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(10)

    src = driver.page_source
    soup = BeautifulSoup(src, 'lxml')

    trs=soup.find_all("tr")
    for tr in trs:
        dict_={}
        tds=tr.find_all('td')
        university=tds[0].text
        study=tds[1].find('b').text
        text=tds[1].text

        dict_['university']=university
        dict_['cluster']=cluster
        dict_['degree']='bachelor'
        dict_['study']=study
        dict_['text']=text
        dict_['url']=url
        dict_list.append(dict_)
    driver.close()

    json_string = json.dumps(dict_list)
    with open(f"../bot_for_recruiters/datas/datas_json/{cluster}.json", "w") as json_file:
        json.dump(dict_list, json_file, indent=4)


json_string = json.dumps(dict_list)
with open(f"../bot_for_recruiters/datas/datas_json/data.json", "w") as json_file:
    json.dump(dict_list, json_file, indent=4)
```
